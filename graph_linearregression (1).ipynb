```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
saldf = pd.read_csv('/content/drive/MyDrive/archive (3) (1).zip')
```

```python
new_saldf = saldf [["Year","Engine Size"]]
print(new_saldf)
```

```python
inp = new_saldf[['Year']]
out = new_saldf['Engine Size']
```

```python
LR = LinearRegression()
```

```python
LR.fit(inp,out)
```

```python
LinearRegression()
```

```python

LR.predict([[2016]])
```

```python
plt.scatter(inp, out, color='blue', label='Original Data')
plt.plot(inp, LR.predict(inp), color='red', label='Regression Line')
plt.xlabel('Year')
plt.ylabel('Engine Size')
plt.title('Linear Regression: Year vs Engine Size')
plt.legend()
plt.show()
```
